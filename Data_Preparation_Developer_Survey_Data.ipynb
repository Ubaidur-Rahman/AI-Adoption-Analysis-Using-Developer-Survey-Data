```python
# Specify the names of the columns to check for all-null values
columns_to_check = ['AISearchHaveWorkedWith', 'AISearchWantToWorkWith',
       'AIDevHaveWorkedWith', 'AIDevWantToWorkWith', 'AISelect',
       'AISent', 'AIAcc', 'AIBen', 'AIToolInterested in Using',
       'AIToolCurrently Using']

# Drop rows where the specified columns have all null values
filtered_survey_data.dropna(subset=columns_to_check, how='all', inplace=True)
```

```python
threshold = (filtered_survey_data.shape[1] / 2)*1.3
threshold
```

```python
null_counts_per_row = filtered_survey_data.isnull().sum(axis=1)
null_counts_per_row
# Delete rows where the count of null values is more than threshold
filtered_survey_data = filtered_survey_data[null_counts_per_row < threshold]
```

```python
filtered_survey_data.shape
```

```python
filtered_survey_data
```

```python
# Extract unique categories from the 'AISearchHaveWorkedWith' column
categories = set(filtered_survey_data['AISearchHaveWorkedWith'].dropna().str.split(';').sum())

# Add "AISearchHaveWorkedWith_" before each category
categories = ['AISearchHaveWorkedWith_' + category for category in categories]

# Create columns for each category and initialize with 0
for category in categories:
    filtered_survey_data[category] = 0

# Iterate through each row and update the category columns accordingly
for index, row in filtered_survey_data.iterrows():
    if pd.notna(row['AISearchHaveWorkedWith']):
        for category in row['AISearchHaveWorkedWith'].split(';'):
            filtered_survey_data.at[index, 'AISearchHaveWorkedWith_' + category] = 1

```
